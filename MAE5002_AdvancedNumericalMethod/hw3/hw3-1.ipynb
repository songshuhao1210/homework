```python
import numpy as np
import matplotlib.pyplot as plt
```

### parameters

```python
flag_err = 1e-12      # error treshold
flag_max_iter = 100
xx = np.arange(-2,2,0.05)
```

### plot the function in comparison with y = x

```python
def func(x,flag_func):
    if flag_func == 1:
        return pow(x,5)-7*pow(x,3)-2*pow(x,2)+2
    elif flag_func == 2:
        return np.sin(np.cos(x))
    elif flag_func == 3:
        return pow(x,2)*np.sin(x+np.pi/3)
    else:
        return pow(x,2)+np.sin(x+np.pi/3)
```

```python
def plot_com(xx,func_this,flag_func):
    plt.figure(figsize = (7,5))
    plt.plot(xx,xx)
    plt.plot(xx,func_this)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(('y = x','y = g(x)'),fontsize=12, numpoints=1,loc=4)
    plt.savefig('figures/hw3-1_comparison_flag_'+str(flag_func)+'.png',dpi=300)
```

```python
for flag_func in range(1,5):
    func_this = np.zeros(np.size(xx))
    for x in xx:
        func_this[xx == x] = func(x,flag_func)
    plot_com(xx,func_this,flag_func)
```

### Bisection method

```python
def func_fix(x,flag_func):
    if flag_func == 1:
        return pow(x,5)-7*pow(x,3)-2*pow(x,2)+2-x
    elif flag_func == 2:
        return np.sin(np.cos(x))-x
    elif flag_func == 3:
        return pow(x,2)*np.sin(x+np.pi/3)-x
    else:
        return pow(x,2)+np.sin(x+np.pi/3)-x
```

```python
def bisec(flag_err,flag_max_iter,x_left,x_right):
    for i in range(1,flag_max_iter):
        x0 = (x_left+x_right)/2
        if func_fix(x0,flag_func)*func_fix(x_right,flag_func)>0:
            x_right = x0
        elif func_fix(x0,flag_func)*func_fix(x_left,flag_func)>0:
            x_left = x0
        else:
            x_right = x0
            x_left = x0
        err = np.abs(x_right-x_left)
        print(f'iteration number={i:d} , error={err:.5E}, x0={x0:.14f}')
        if err < flag_err:
            return x0
    print('iteration reaches the limitation')
    return x0
```

```python
def plot_com_1(xx,func_this,flag_func,x0,flag_num):
    plt.figure(figsize = (7,5))
    plt.plot(xx,xx)
    plt.plot(xx,func_this)
    plt.plot(x0,func(x0,flag_func),'o')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(('y = x','y = g(x)','fix point'),fontsize=12, numpoints=1,loc=4)
    plt.savefig('figures/hw3-1_fix_point_flag_'+str(flag_func)+'_'+str(flag_num)+'.png',dpi=300)
```

### (a)

```python
flag_func = 1
```

```python
x_left = 0
x_right = 1
x0 = bisec(flag_err,flag_max_iter,x_left,x_right)
```

```python
func_this = np.zeros(np.size(xx))
for x in xx:
    func_this[xx == x] = func(x,flag_func)
plot_com_1(xx,func_this,flag_func,x0,1)
```
